**Example 2:** The reaction time x (in minutes) of a certain chemical process follows a uniform probability distribution with 5$\leq$x$\leq$10.

**(a)** Draw the graph of the density curve.



```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution
min_reaction_time = 5  # Minimum reaction time in minutes
max_reaction_time = 10  # Maximum reaction time in minutes

# Create a range of reaction times
reaction_times = np.linspace(min_reaction_time, max_reaction_time, 1000)

# Calculate the probability density function (PDF) for the uniform distribution
pdf = np.full_like(reaction_times, 1 / (max_reaction_time - min_reaction_time))

# (a) Draw the graph of the density curve
plt.plot(reaction_times, pdf, label='PDF')
plt.fill_between(reaction_times, pdf, color='green', alpha=0.2, label='Area')
plt.title('Uniform Probability Distribution')
plt.xlabel('Reaction Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

```

**(b)** What is the probability that the reaction time is between 6 and 8 minutes?

```python
probability_between_6_and_8 = (8 - 6) / (max_reaction_time - min_reaction_time)
print(f'(b) Probability that reaction time is between 6 and 8 minutes: {probability_between_6_and_8:.2f}')
```

**(c)** What is the probability that the reaction time is less than 8 minutes?

```python
probability_less_than_8 = (8 - min_reaction_time) / (max_reaction_time - min_reaction_time)
print(f'(c) Probability that reaction time is less than 8 minutes: {probability_less_than_8:.2f}')
```

**(d)** What is the probability that the reaction time is exactly 7 minutes?

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution
min_reaction_time = 5  # Minimum reaction time in minutes
max_reaction_time = 10  # Maximum reaction time in minutes
target_time = 7  # Target time in minutes

# Calculate the probability density function (PDF) for the uniform distribution
pdf = 1 / (max_reaction_time - min_reaction_time)

# Define a small range around the target time (e.g., within 1e-6 of 7 minutes)
small_range = 1e-6

# (a) Draw the graph of the density curve
plt.plot([min_reaction_time, max_reaction_time], [pdf, pdf], label='PDF')
plt.fill_between([min_reaction_time, max_reaction_time], [pdf, pdf], color='blue', alpha=0.2, label='Area')
plt.title('Uniform Probability Distribution')
plt.xlabel('Reaction Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# (d) What is the probability that the reaction time is exactly 7 minutes?
probability_exact_7 = pdf * small_range
print(f'(d) Probability that reaction time is exactly 7 minutes: {probability_exact_7:.2e}')
```

**(e)** Find the 20th percentile.

```python
# Parameters for the uniform distribution
min_reaction_time = 5  # Minimum reaction time in minutes
max_reaction_time = 10  # Maximum reaction time in minutes
percentile = 0.2  # 20th percentile

# (e) Find the 20th percentile
percentile_value =  (max_reaction_time - min_reaction_time) * percentile
area = percentile_value + min_reaction_time

print(f'(e) 20th percentile: {area:.2f} minutes')
```

#Application of the Normal Distribution

**Example 1:** Scores on the Stanford-Binet Intelligence Test are normally distributed with mean =10 and standard deviation =16 .

**(a)** People with IQ scores above 132 are considered to be gifted. What percent of people are gifted?

```python
# (a) IQ score above 132
from scipy.stats import norm
mean = 100
stdv = 16
x = 132

prob = 1 - norm.cdf(x, loc = mean, scale=stdv)
print("Probability of p(X >",x,")=",prob)
```

**(b)** What is the probability that a randomly selected individual has an IQ between 90 and 120?

```python
# (b) IQ score between 90 and 120
from scipy.stats import norm
mean = 100
stdv = 16
a = 90
b = 120

prob = norm.cdf(b, loc = mean, scale=stdv) - norm.cdf(a, loc = mean, scale=stdv)
print("Probability of p(",a ,"< X >",b,")=",prob)
```

**(c)** Find the percentile ranking of an individual whose IQ is 125.

```python
# (c) IQ score less than
from scipy.stats import norm
mean = 100
stdv = 16
x = 125

prob = norm.cdf(x, loc = mean, scale=stdv)
print("Probability of p( X <",x,")=",prob)
```

#Sampling Distribution of the Sample Proportion, $\hat{p}$



**Example 5:**  According to credicard.com, the proportion of adults who do not own a credit card is 0.29.  Suppose a random sample of 500 adults is asked, Do you own a credit card?

**(a)**  What is the probability that in a random sample of 500 adults, between 25% and 30% do not own a credit card?



```python
from scipy.stats import norm
import math

mu_phat = .29
sigma_phat = math.sqrt(.29*(1-.29)/500)

norm.cdf(.30,mu_phat,sigma_phat) - norm.cdf(.25,mu_phat,sigma_phat)
```

**(b)**  Would it be unusual for a random sample of 500 adults to result in 125 or fewer who do not own a credit card?  Why?

```python
from scipy.stats import norm
import math

mu_phat = .29
sigma_phat = math.sqrt(.29*(1-.29)/500)
x=125
n=500
phat=x/n

norm.cdf(phat,mu_phat,sigma_phat)
```

**Practice 2:**  The Pew Research Center recently reported that 18% of women 40-44 years old have never given birth.  Suppose a random sample of 250 adult women aged 40-44 years old is taken.

**(a)**  What is the probability that in a random sample of 250 women aged 40-44 years less than 10% have never given birth?

np(1-p)$\geq$10

$\mu_\hat{p}$=0.18

```python
from scipy.stats import norm
import math

mu_phat = .18
sigma_phat = math.sqrt(.18*(1-.18)/250)


norm.cdf(0.10,mu_phat,sigma_phat)
```
