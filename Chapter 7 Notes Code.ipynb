# Finding Probability for Normal Distribution

**Example 2:** Find the requested probability for the standard normal distribution and sketch the represented area.

**(a) P(-1.3 < z < 2.3)**

```python
#Import Library
from scipy.stats import norm

#Find prob
norm.cdf(2.3)-norm.cdf(-1.3)
```

**Alternate Python Code**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def area_under_normal_curve(mean, std_dev, lower_bound, upper_bound):
    """
    Calculate the area under the normal curve between lower_bound and upper_bound.

    Parameters:
    - mean: Mean of the normal distribution
    - std_dev: Standard deviation of the normal distribution
    - lower_bound: Lower bound
    - upper_bound: Upper bound

    Returns:
    - Area under the normal curve between lower_bound and upper_bound
    """
    area = norm.cdf(upper_bound, loc=mean, scale=std_dev) - norm.cdf(lower_bound, loc=mean, scale=std_dev)
    return area

def plot_normal_distribution(mean, std_dev, lower_bound, upper_bound):
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
    y = norm.pdf(x, loc=mean, scale=std_dev)

    plt.plot(x, y, 'k-', linewidth=2)
    plt.fill_between(x, y, where=[(lower_bound <= xi <= upper_bound) for xi in x], color='red', alpha=0.4)

    plt.title('Normal Distribution')
    plt.xlabel('X')
    plt.ylabel('Probability Density Function')
    plt.show()

# Prompt user for input
mean_value = float(input("Enter the mean of the normal distribution: "))
std_deviation = float(input("Enter the standard deviation of the normal distribution: "))
lower_limit = float(input("Enter the lower bound: "))
upper_limit = float(input("Enter the upper bound: "))

result = area_under_normal_curve(mean_value, std_deviation, lower_limit, upper_limit)
print(f"Area under the normal curve between {lower_limit} and {upper_limit}: {result}")

# Plotting the normal distribution with shaded area
plot_normal_distribution(mean_value, std_deviation, lower_limit, upper_limit)
```

# Application

**Example 5:** A pediatrician obtains the heights of her 200 three-year-old female patients. The heights are approximately normally distributed, with mean 38.72 inches and standard deviation 3.17 inches.

**(a)** What’s the proportion of the 3-year-old females that have a height less than 35 inches?

```python
#Import Library
from scipy.stats import norm

#Find prob
norm.cdf(35, 38.72, 3.17)
```

**(b)** What’s the probability that a randomly selected 3-year-old will have a height more than 40 inches?

```python
#Import Library
from scipy.stats import norm

#Find prob
1-norm.cdf(40, 38.72, 3.17)
```

**(c)** What’s the probability of randomly selecting a 3-year-old that has a height between 37 and 39 inches?

```python
#Import Library
from scipy.stats import norm

#Find prob
norm.cdf(39, 38.72, 3.17)-norm.cdf(37, 38.72, 3.17)
```

**Example 6:** The scores earned on a national standardized exam are approximately normally distributed with mean score of 516 and standard deviation 116. What scores separate the middle 90% of test takers from the bottom and top 5%? In other words, find the 5th and 95th percentiles.

```python
#Import Library
from scipy.stats import norm

#Find each score
x1 = norm.ppf(.05,516,116)
x2 = norm.ppf(.95, 516, 116)

print("The 5th percentile is", x1)
print("The 95th percentile is", x2)
```

# Sampling Distribution of Proportions

**Example 1:**  The National Coffee Association reported that 63% of U.S. adults drink coffee daily. A random sample of 250 U.S. adults is selected.

**(b)** What’s the probability that more than 67% of U.S. adults drink coffee daily?

```python
#Import Libraries
from scipy.stats import norm
import math

#Identify center and spread
mu_phat = .63
sigma_phat = math.sqrt(.63*(1-.63)/250)

#Find prob
1-norm.cdf(.67,mu_phat,sigma_phat)
```

**(c)** Would it be unusual for 140 or fewer adults to drink coffee?

```python
#Import Libraries
from scipy.stats import norm
import math

#Identify center and spread
mu_phat = .63
sigma_phat = math.sqrt(.63*(1-.63)/250)
phat = 140/250

#Find prob
norm.cdf(phat,mu_phat,sigma_phat)
```
