## 2. The Regression Line

The least squares regression line is a line that best fits the data and and is also the line that minimizes the distance of the data points to the line.

**Question 2.1:** Find the line of best fit for the old faithful example.

```python
#Importing Library
import matplotlib.pyplot as plt

# Fit a linear regression line using numpy
slope, intercept = np.polyfit(x, y, 1)

#Output the linear regression equation
equation = f"Regression Equation: y = {slope:.4f}x + {intercept:.4f}"
print(equation)

# Plot the scatter plot and regression line
plt.scatter(x, y, label='Data points')
plt.plot(x, slope * x + intercept, color='red', label='Regression line')

plt.title('Scatter Plot with Regression Line')
plt.xlabel('Variable') #Put the name of the indpendent variable
plt.ylabel('Variable 2') #Put the name of the dependent variable
plt.legend()
plt.show()
```
